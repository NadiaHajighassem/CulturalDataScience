```python
# import packages
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load in data
df = pd.read_csv('mp_data1.csv') 
```

```python
# lets check it out
df.head(30)
```

```python
# data cleaning
df_clean = df[df['sender'] == 'target']

# lets check
df_clean.head(20)
```

```python
# lets check if target is the only unique item in the column sender
set(df_clean['sender'])
```

```python
# clean the column
df_clean.loc[:, 'response_action'] = df_clean['response_action'].str.replace(r'keypress\(|\)', '', regex=True)
df_clean.loc[:, 'correct'] = df_clean['response_action'] == df_clean['correct_response']
```

```python
list(df_clean)
```

```python
# selecting specific columns that we want
data = df_clean.loc[:, ['ID', 'duration', 'condition', 'soa_condition','correct', 'congruence']]
data.head(5)
```

```python
# remove practice from condition column
data = data[data['condition'] != 'practice']
# remove incorrect answers
data = data[data['correct'] == True]

# lets check if correct is only there
data.correct.unique()
```

```python
sns.pointplot(data=data, x='soa_condition', y='duration', hue= 'congruence', order=['short', 'long'])
plt.ylabel('RT(ms)')
plt.xlabel('SOA condition')
plt.title('My awesome plot')
```
